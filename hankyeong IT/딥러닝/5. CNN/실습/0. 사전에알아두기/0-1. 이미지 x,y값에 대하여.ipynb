### 이미지 정규화(이미지스케일링: 특정한 범위내로 조정)

```python
from PIL import Image
import numpy as  np
import matplotlib.pyplot as plt

img=Image.open('dog.jpg')
img=img.resize((180,180))  # 작업의 편의성을 위해 임의로 사이즈를 조정
imgArr=np.array(img)

숫자범위0에서255=imgArr
숫자범위0에서1=imgArr/255
숫자범위마이너스1에서1=imgArr/127-1    # 진한색은더 진하게, 밝은색은 더 밝게 (티처블머신러닝 스케일링방법임)

def info(var):
    print(np.min(var),'~', np.max(var))
    plt.imshow(var)
    plt.show()
    
info(숫자범위0에서255)
info(숫자범위0에서1)
info(숫자범위마이너스1에서1)
```

### numpy 자료 이미지로 저장하기
- array를 이미지로 저장하는 명령을 실행할때 0~255로 저장하는것을 권장함.
    - 0~1, 또는 -1~1로 스케일링(정규화) 된 자료는 다시 0~255로 변경해야함.
    - 0~1 또는 -1~1로 스케일링 된 자료는 검정색(0)으로 저장될수도 있음. 이미지저장명령에 따라서 알아서 0~255로 변환하기도함
- 문제는 0~255자료를 어떤 방식으로 스케일링(정규화) 했는지를 알고 있어야함

```python
import imageio
imageio.imwrite('넘파이0에서255.jpg', 숫자범위0에서255)
Image.open('넘파이0에서255.jpg')
```

```python
import imageio
imageio.imwrite('넘파이숫자범위0에서1.jpg', 숫자범위0에서1)
img=Image.open('넘파이숫자범위0에서1.jpg')
np.min(np.array(img)),'~', np.max(np.array(img)), np.array(img)[:,:,0]     # 출력해보면 0~1사이값이 0~ 255값으로 변경됨
```

```python
import imageio
imageio.imwrite('넘파이숫자_숫자범위마이너스1에서1.jpg', 숫자범위마이너스1에서1)
img=Image.open('넘파이숫자_숫자범위마이너스1에서1.jpg')
plt.imshow(img)

np.min(np.array(img)),'~', np.max(np.array(img)), np.array(img)[:,:,0]     # 출력해보면 0~1사이값이 0~ 255값으로 변경됨
```
