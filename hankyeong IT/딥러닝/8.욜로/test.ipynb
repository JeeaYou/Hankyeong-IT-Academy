```python
# 1자리 숫자를 2자리로 만들기
folder='./test/'
elseWord='test_'
num=len(elseWord)
extendName='.txt'

fileList=os.listdir(folder)
# test부터 .txt전까지
tmp=[len(x[num:-4]) for x in fileList]
for x in fileList:
    # 숫자자리수를 max(tmp)만큼 만든다
    newFileName=elseWord+str(x[num:-4]).zfill(np.max(tmp)) + extendName
    os.rename(folder+x, folder+newFileName)
```

```python
# 목록리스트를 불러오고 텍스트의 [0]을 불러오기

from glob import glob
import numpy as np
import matplotlib.pyplot as plt

fileList = glob('./BTS_MV_YOLO/labels/*.txt')
videoInfo = []

for x in fileList:
    f = open(x,'r')
    lines = f.readlines()
    for line in lines:
        videoInfo.append([])
        videoInfo[-1].append(line.split(' ')[0])
        videoInfo[-1].append(x[-7:-4])
videoInfo = np.array(videoInfo)
videoCount = np.unique(videoInfo[:,0], return_counts = True)
plt.barh(videoCount[0],videoCount[1])
```
