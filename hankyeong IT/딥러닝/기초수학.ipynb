```python
import math
import matplotlib.pyplot as plt
import numpy as np

a = [6,13]
b = [17,8]


plt.figure(figsize=(10,10))
plt.xlim(0, 20)          
plt.ylim(0, 20)
plt.xticks(ticks=np.arange(0, 20, step=1))
plt.yticks(ticks=np.arange(0, 20, step=1))

plt.arrow(0, 0, a[0], a[1], head_width = .5, head_length = .5, color = 'red')
plt.arrow(0, 0, b[0], b[1], head_width = .5, head_length = .5, color = 'blue')
plt.arrow(a[0], a[1], 11, -5, head_width = .5, head_length = .5, color = 'green')
plt.arrow(6,8,0,5, head_width = .5, head_length = .5, color = 'pink')
plt.arrow(6,8,11,0, head_width = .5, head_length = .5, color = 'black')

plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.title("Vector")
plt.grid()
plt.show()
```

# 삼각함수의 함수개형 및 크기구간

```python
import math
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib import rc
rc('font', family='New Gulim')  #한글처리 

x = np.arange  ( start = -math.pi*2, stop = math.pi*2, step = 0.01)  #x좌표 


plt.title("삼각함수")
plt.plot( x, np.sin(x),'r-' ,label='sin')
plt.xlim(-math.pi*2, math.pi*2)          
plt.ylim(-2, 2)
plt.xticks(ticks=np.arange(-6, 6, step=1))
plt.yticks(ticks=np.arange(-2, 2, step=1))
plt.axis([-math.pi*2, math.pi*2, -2, 2])
plt.legend()
plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.grid()
plt.show()

```

```python
import math
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib import rc
rc('font', family='New Gulim')  #한글처리 

x = np.arange  ( start = -math.pi*2, stop = math.pi*2, step = 0.01)  #x좌표 


plt.title("삼각함수")
plt.plot( x, np.cos(x),'b-' ,label='blue')
plt.xlim(-math.pi*2, math.pi*2)          
plt.ylim(-2, 2)
plt.xticks(ticks=np.arange(-6, 6, step=1))
plt.yticks(ticks=np.arange(-2, 2, step=1))
plt.axis([-math.pi*2, math.pi*2, -2, 2])
plt.legend()
plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.grid()
plt.show()

```

```python
import math
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib import rc
rc('font', family='New Gulim')  #한글처리 

x = np.arange  ( start = -math.pi*2, stop = math.pi*2, step = 0.01)  #x좌표 


plt.title("삼각함수")
plt.plot( x, np.tan(x),'g-' ,label='tangent')
plt.xlim(-math.pi*2, math.pi*2)          
plt.ylim(-20, 20)
plt.xticks(ticks=np.arange(-6, 6, step=1))
plt.yticks(ticks=np.arange(-20, 20, step=1))
plt.axis([-math.pi*2, math.pi*2, -2, 2])
plt.legend()
plt.axhline(0, color='gray', alpha = 0.1)
plt.axvline(0, color='gray', alpha = 0.1)
plt.grid()
plt.show()
```

삼각함수의 역함수 : 삼각함수 값을 각도로 변환하는 함수
