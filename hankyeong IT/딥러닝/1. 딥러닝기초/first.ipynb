```python
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    return np.array(x > 0, dtype=np.int)
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = step(x)
plt.title('Step Function')
plt.plot(x,y)
plt.show()
```

```python
# 시그모이드 함수 그래프를 그리는 코드
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu Function')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```

```python
# 임의 x,y값을 제작
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
```

```python
# wx+b의 선형(일차방정식)선의 공식은 존재하지만 인공지능은 그 공식(통계학적 공식)을 무시하고 
# 무작위로 생성된 w,b 값을 갖고 계산하고 ,오차줄을 반복하느 과정
w = 0.2 ; b = 0.1
y_hat = w*x+b
plt.scatter(x,y)
plt.plot(x,y_hat)

# 오차는 얼마인지 
MSE = np.sum(y-y_hat)**2
plt.title(MSE)
```

```python
# w,b 값을 계속 갱신하면서 mse를 확인해보겠음
# 임의로 w,b값을 갱신해보겠음
wData =  [0.2,0.1,0.3,2]
bData = [0.1,0.1,0,-0.5]
for w,b in zip(wData,bData):
    y_hat = w*x+b
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    # 오차는 얼마인지 
    MSE = np.mean((y-y_hat)**2)
    plt.title(MSE)
    plt.show()
```

```python
# w,b 값을 계속 갱신하면서 mse를 확인해보겠음
# 임의로 w,b값을 갱신해보겠음
plt.figure(figsize=(20,6))
wData =  [0.2,0.1,0.3,2]
bData = [0.1,0.1,0,-0.5]
cnt = 0

for w,b in zip(wData,bData):
    cnt+=1
    plt.subplot(1,5,cnt)
    y_hat = w*x+b
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    # 오차는 얼마인지 
    MSE = np.mean((y-y_hat)**2)
    plt.title(MSE)
plt.show()
```

```python
# w,b 값을 계속 갱신하면서 mse를 확인해보겠음
# 임의로 w,b값을 갱신해보겠음
import errno
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.figure(figsize=(20,6))
w = 2 ; b = 0.1
cnt = 0

for xx in range(5):
    cnt+=1
    plt.subplot(1,5,cnt)

    y_hat = w*x+b

    plt.ylim([-8,8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    # 오차는 얼마인지 
    err = np.mean((y-y_hat)**2)
    plt.title(err)

    w = w-w*err*0.01
    b = b-b*err*0.01
plt.show()
```

```python
### w,b값을 계속 갱신하면서 mse를 확인해보겠음
### 임의로 w,b값을 갱신해보겠음
x=np.random.randn(1000)
y=np.random.randn(1000)

plt.figure(figsize=(20,6))
cnt=0
# 웨이트값이 너무 크면 에러도 커진다
w=2 ; b=0.1
# 러닝 레이트
lr=0.0001
# 
epochs=10

for xx in range(epochs): 
    cnt+=1
    plt.subplot(1,epochs,cnt)

    y_hat=w*x+b
    
    plt.ylim([-8,8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    #  에러 값을 만들기
    err=np.mean((y-y_hat)**2)
    plt.title(err)
    
    # w와b값을 갱신
    w=w-w*err*lr
    b=b-b*err*lr
plt.show()
```

```python
### w,b값을 계속 갱신하면서 mse를 확인해보겠음
### 임의로 w,b값을 갱신해보겠음
x=np.random.randn(1000)
y=np.random.randn(1000)

plt.figure(figsize=(20,6))
cnt=0
# 웨이트값이 너무 크면 에러도 커진다
# 웨이트와 바이어스의 갱신은 훈련데이터만 갖고 한다
# 웨이트와 바이어스의 갱신은 오차가 작으면 변화율이 작음.
# 과적합(오버피팅)은 웨이트와 바이어스의 갱신은 잘 안이루어지는데 (훈련데이터에 잘 맞음) 
# 테스트 데이트는 잘 안맞음
w=2 ; b=0.1
# 러닝 레이트
lr=0.0001
# 
epochs=10

for xx in range(epochs): 
    cnt+=1
    plt.subplot(1,epochs,cnt)

    y_hat=w*x+b
    
    plt.ylim([-8,8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    #  에러 값을 만들기
    err=np.mean((y-y_hat)**2)
    plt.title(err)
    
    # w와b값을 갱신
    w=w-w*err*lr
    b=b-b*err*lr
plt.show()
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(1, activation='relu'),
  tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 기본제공되는 32개의 배치단위로 w,b를 갱신하고 1에 포크가 끝나면
# 그 w,b를 validation_data에 넣어서 검증함
# 오버피팅(과적함):훈련데이터의 loss는 점점작아지는데 ,validation_data loss는 점점 커지는지점
model.fit(x_train, y_train,
            validation_data=(x_test,y_test),epochs=5)
```
