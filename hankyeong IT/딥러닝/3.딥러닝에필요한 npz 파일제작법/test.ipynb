```python
# 풀더명을 csv로 제작
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from glob import glob

folderName = glob('./clothes_dataset/*')
dataList = []

for folder in folderName:
    fileList = glob( folder + './*jpg') 
    for file in fileList:

# file = './clothes_dataset\\white_shorts.\\fd76a8fe7e6c5d2fe5f346c3207f75325a781d82.jpg'

        fileName = file
        colorName = file.split('\\')[1].split('_')[0]
        categoryName = file.split('\\')[1].split('_')[-1].replace('.','')
        
        dataList.append({'Name':fileName, 
                        'colorName':colorName, 
                        'categoryName':categoryName})
df = pd.DataFrame(dataList)
df
```

```python
# 원핫인코딩
pd.get_dummies(df[['colorName','categoryName']])
```

```python
# 원핫인코딩 
df1 = pd.get_dummies(df, columns=['colorName','categoryName'])     # 모든 컬럼스 출력
df1.columns=['Name', 'black', 'blue', 'brown','green', 'red', 'white','dress', 'pants', 'shirt','shoes', 'shorts']
df1.to_csv('./train.csv')  
df1.to_csv('./val.csv')  
df1
```
