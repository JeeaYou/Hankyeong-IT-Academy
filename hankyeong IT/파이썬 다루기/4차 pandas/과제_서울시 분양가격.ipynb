```python
df.columns=['지역명', '규모구분', '연도', '월', '분양가격']
```

```python
df.head()
```

```python
#df['분양가격']=='  '  None 채우기
df.loc[df['분양가격']=='  ','분양가격']=None
```

```python
df.isnull().sum()
```

```python
#분양가격 ',' 없애기

df['분양가격']=df['분양가격'].str.replace(',','')
```

```python
#'분양가격']=='-' 에 None 입력
df.loc[df['분양가격']=='-','분양가격']=None
```

```python
# 분양가격 notnull 표를 따로 저장함
df_not_null=df[df['분양가격'].notnull()]
```

```python
df_not_null['분양가격']=df_not_null['분양가격'].astype('int')
```

```python
df_not_null.dtypes
```

```python
# 1. df 분양가격에서...  isnull만추출
# 최종미션: df분양가격의 isnull을  연도별, 지역별,규모별, 분양가격 평균으로 채우기
```

```python
df_null=df[df['분양가격'].isnull()]
```

```python
df_not_null
```

```python
df_null.head()
```

```python
df_res=pd.concat([df_not_null,df_null])     #여기까지 데이터 전처리
```

```python
df_res
```

```python
df_res.columns
```

```python
df_res=df_res.sort_index()
```

```python
df_g=df_res.groupby(['연도','지역명','규모구분'])
```

```python
list(df_g)
```

```python
res=[]
for i , data in df_g:
    data=data.fillna(data.mean())
    res.append(data)
```

```python
len(df_g)
```

```python
len(res)
```

```python
df_완성=pd.concat(res)
```

```python
# 연도별,지역별, 규모구분별 분양가 평균가격을 분양가 null에 채우기
df_g=df_res.groupby(['연도','지역명','규모구분'])

res=[]
for i,data in df_g:

    data=data.fillna(data.mean())
    res.append(data)
df_완성=pd.concat(res)
df_완성=df_완성.sort_index()
```
