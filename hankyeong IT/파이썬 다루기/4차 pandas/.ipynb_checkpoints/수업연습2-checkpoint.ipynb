```python
def max_o(s_df,x):     #x는 열이 들어간다
    mn,c,mx=s_df[x].quantile([0.25,0.5,0.75])
    IRQ=mx-mn
    min_outlier=mn+(IRQ*1.5)
    return df1[s_df[x]<min_outlier]
```

```python
def min_o(s_df,x):     #x는 열이 들어간다
    mn,c,mx=s_df[x].quantile([0.25,0.5,0.75])
    IRQ=mx-mn
    max_outlier=mx+(IRQ*1.5)
    return df1[s_df[x]>=max_outlier]
```
