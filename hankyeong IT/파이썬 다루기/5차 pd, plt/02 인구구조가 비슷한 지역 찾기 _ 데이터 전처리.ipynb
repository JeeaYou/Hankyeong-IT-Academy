####  1-2. column명 바꾸기

```python
# culumn 명 확인
df.columns
#display(df.head(3))
```

```python
# 컬럼명을 '_'로 나눈훈 마지막 문자만 컬럼명으로 사용하기 (예: '2019년02월_계_총인구수' => '총인구수')
# 문자열  split()함수 사용
df.columns=[i.split('_')[-1] for i in df.columns]
df.head(3)
```

#### 1-3 전체 데이터를 총 인구수로 나누어 비율로 변환
1. 전체 데이터를 숫자로 바꾸기
2. 전체 데이터를 총 인구수(df['총인구수']) 나눔


* 1. 전체 데이터를 숫자로 바꾸기 

```python
# 전체 데이터를 숫자로 바꾸기(1)

#(replace함수 활용)  => (, 없애기 예:53,738 => 53738) 

# 방법1
df_1=df.copy()

for i in df_1:
    try:
        df_1[i]=df_1[i].str.replace(',','')
        df_1[i]=df_1[i].astype('int64')
    except:
        pass
df_1.head(3)
```

```python
# 전체 데이터를 숫자로 바꾸기(3)
# 방법2 : for문 활용
df_1=df.copy()
for i in df_1:
    df_1[i]=[int(str(j).replace(',','')) for j in df_1[i]]

df_1.head(3)
```

```python
# 전체 데이터를 숫자로 바꾸기(3)
#방법3 :함수 정의, apply함수 활용

def stoi(x):
    res=[int(str(i).replace(',','')) for i in x]
    return res
df_1=df_1.apply(stoi)
df_1.head(3)
```

```python
# 전체 데이터를 숫자로 바꾸기(4)
#방법3 :apply, lambda 함수 활용
df_1=df.copy()
df_1=df_1.apply(lambda x: [int(str(i).replace(',','')) for i in x])

df_1.head(3)
```

* 3) 전체 데이터를 총 인구수(df['총인구수']) 나눔
* 4) 총인구수와 연령구간인구수 삭제

```python
df_1=df_1.div(df_1['총인구수'], axis=0)
```

```python
#총인구수, 연령구간인구수 삭제
# del df_1['총인구수'], df_1['연령구간인구수'] 
df_1=df_1.iloc[:,2:]
df_1.head(3)
```
