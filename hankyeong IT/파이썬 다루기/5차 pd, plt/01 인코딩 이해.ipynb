### 1-2 파이썬의 ord와 chr함수 (5분)

- chr함수 : chr(숫자)를 이용해서 숫자에 해당하는 문자를 반환
- ord함수 : ord (문자)를 이용해서 해당 문자에 해당하는 유니코드 정수가 잘 반환



```python
#ord(문자)

print(ord('A'))
print(ord('a'))
print(ord('*'))
print(ord('\n'))
print(ord('\t'))
```

```python
# chr(숫자)
print(chr(65))
print(chr(97))
print('안녕하세요' + chr(10) + '홍길동입니다.')
```

### 1-3: 리스트 for, 컴프리헨션, 함수

```python
#각각의 데이터를 유니코드로 변환
data=['a','b','c','A','B','C','가','나','ㄴ','A','B', '%', '/']
for i in data:
    print(i,'를 컴에서는 어떻게 인식할까? ==>', ord(i))
```

```python
#리스트 컴프리헨션
data=['a','b','c','A','B','C','가','나','ㄴ','A','B', '%', '/']
[('i','를 컴에서 어떻게 인식힐까?',ord(i)) for i in data]
```

```python
# 소문자를 대문자로 바꾸기(1) :for
tmp='abc'
for i in tmp:
    print(chr(ord(i)-32))
```

```python
# 소문자를 대문자로 바꾸기(2) :append
tmp='abc'
res=[]
for i in tmp:
    res.append(chr(ord(i)-32))
print(res)
```

```python
# 소문자를 대문자로 바꾸기(3) : 리스트 컴프리헨션
tmp='abc'
res=[chr(ord(i)-32)for i in tmp]
print(res)
```

```python
# 소문자를 대문자로 바꾸기(4) : 함수

def myUpper(x):
    return chr(ord(x)-32)

tmp='abc'

print(list(map(myUpper,tmp)))

[myUpper(i) for i in tmp]
```

```python
# 소문자를 대문자로 바꾸기(4)기본 내장함수 사용: upper(), lower()
print('abc'.upper())
print('ABC'.lower())
```

### 1-4: 딕셔너리 구조와 이중for



```python
#딕셔너리 구조 (for)

data={'category':['영어','기호','한글'],
     'dat':[['A','B','a','b'],['/','%'],['가','나']]}

dt=[]
for key,v in zip(data['category'],data['dat']):
    for j in v:
        dt.append([key,j])
pd.DataFrame(dt,columns=['구분','값'])
```

```python
#딕셔너리 구조 (리스트 컴프리헨션)

data={'category':['영어','기호','한글'],
     'dat':[['A','B','a','b'],['/','%'],['가','나']]}
dt=[[key,j] for key,v in zip(data['category'],data['dat']) for j in v]
pd.DataFrame(dt,columns=['구분','값'])
```
