### 연령대별 인구 현황 분석(예제: age_2.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서  음수 부호설정
plt.rc('axes', unicode_minus=False)
```

```python
df=pd.read_csv('./age_2.csv', encoding='cp949')
```

```python
# 1. 지역을 index로 설정
df=df.set_index('지역')
```

```python
df.head(3)
```

```python
# 2. 마지막 행정구역코드을 제외함
df=df.iloc[:,:-1]
```

```python
# 3. 열 5개씩 추출하여 행방향으로 합을 구하고
#    구한 합만 가지고 데이터프레임을 작성함
#    이때, 컬러명도 새로 지정함 (예: 0-4세, 5-9세, 10-14세.. 등)
res=df.iloc[:,:5].sum(axis=1)

columns=['0-4세']
for i in range(5,102,5):
    columns.append(f'{i}-{i+4}세')
    tmp=df[df.columns[i:i+5]].sum(axis=1)
    res=pd.concat([res,tmp],axis=1)
```

```python
# 4. 컬럼명을 지정함
res.columns=columns
res
```

```python
# 5. 시도, 시군구 분리
res_1=res.copy()
res_1=res_1.reset_index()
res_1['시도']=res_1['지역'].str.split().str[0]
res_1['시군구']=res_1['지역'].str.split().str[1]
```

```python
res_1
```

```python
#시도데이터만 추출 (시군구가 Nan 데이터만 추출)
res_si=res_1[res_1['시군구'].isnull()]
```

```python
#시도를 index로 설정
res_si=res_si.set_index('시도')
```

```python
#시도별 연령대를 차트로 작성함
res_si.T[1:].plot(figsize=(20,5))
```

```python
#0유아기 비율이 가장 큰 시도는?
res_si['0-4세'].sort_values(ascending=False)[1:6]
```

```python
#0유아기 비율이 가장 큰 시도는 (차트)?
res_si['0-4세'].sort_values(ascending=False)[1:6].plot(kind='bar')
```

```python
#0유아기 비율이 가장 큰 시도는?
res_si['0-4세'].sort_values(ascending=False)[1:6].plot(kind='pie',autopct='%1.1f%%')
```

```python
#100세의 비율이 가장 큰 도시는?
res_si['100-104세'].sort_values(ascending=False)[:5].plot(kind='pie',autopct='%1.1f%%')
```
