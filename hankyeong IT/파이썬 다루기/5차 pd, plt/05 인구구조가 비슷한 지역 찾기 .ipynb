### 인구구조가 비슷한 지역 5곳 찾아 시각화하기 (P268) 예제파일: age_2.csv
1. 데이터 전처리
2. 특정 지역을 입력받고 필터링
3. 특정 지역과 원본데이터의 거리 측정(유사도 측정)

  > - (특정지역-원본데이터)의 제곱의 합이 가장 작은 것  => 가장 유사한 지역 <br>
  > -  유사한 지역 5곳 찾기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서  음수 부호설정
plt.rc('axes', unicode_minus=False)
```

### 1. 데이터 전처리


```python
df=pd.read_csv('age_2.csv',encoding='cp949')
```

```python
df1=df.set_index('지역').iloc[:,:-1]
```

```python
df1.head(3)
```

### 2. 특정 지역을 입력받고 필터링

```python
name='신도림'
df2=df1[df1.index.str.contains(name)]
```

```python
df2.T.plot(figsize=(20,5))
```

### 3. 특정 지역과 원본데이터의 거리 측정(유사도 측정)

  > - (특정지역-원본데이터)의 제곱의 합이 가장 작은 것  => 가장 유사한 지역 <br>
  > -  유사한 지역 5곳 찾기

```python
#(특정지역-원본데이터)
a=df1.sub(df2.iloc[0],axis=1)
```

```python
a
```

```python
b=a**2
```

```python
b
```

```python
# 가장 비슷한 지역
i=b.sum(axis=1).sort_values()[:2].index
df1.loc[i].T.plot(figsize=(20,5))
```

```python
##### 유사한 지역 5곳 찾기
# 가장 비슷한 지역
i=b.sum(axis=1).sort_values()[:6].index
df1.loc[i].T.plot(figsize=(20,5))
```
