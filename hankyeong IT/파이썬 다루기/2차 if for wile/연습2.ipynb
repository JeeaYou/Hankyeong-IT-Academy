```python
import csv
with open('./seoul.csv','r')as f:
    data=csv.reader(f)
    header=next(data)
    max_tmp=-9999999
    max_day=''
    for row in data:
        if row[-1]!='':
            tmp=float(row[-1])
            if max_tmp<tmp:
                max_tmp=tmp
                max_day=row[0]
print(max_day,max_tmp)
```

```python
#기상청이 생긴이후 내 생일 날짜가 가장 낮은 온도는?
import csv
with open('./seoul.csv','r')as f:
    data=csv.reader(f)
    header=next(data)
    min_tmp=9999999
    b_day='-05-03'
    min_day=''
    for row in data:
        if b_day ==row[0][5:]:
            try:
                tmp=float(row[-2])   #숫자float로 설정했을 때 출력값이 오류가 나지 않는다
                if min_tmp>tmp:
                    min_tmp=tmp
                    min_day=row[0]
            except:
                pass
print(min_day,min_tmp)
```

```python
import csv
with open('./seoul.csv','r')as f:
    data=csv.reader(f)
    header=next(data)
    month_8='-08-'
    august=[]
    max_temp=99999
    for row in data:
        if month_8 ==row[0][5:7]:
            try:
                temp=float(row[-1])
                if max_temp>temp:
                    max_temp=temp
                    august=row[0]
            except:
                pass
print(august,max_temp)
```

```python
#기상청이 생긴이후 내 생일 날짜가 가장 낮은 온도는?
with open('./seoul.csv','r')as f:
    data=csv.reader(f)
    header=next(data)
    month_8=[]
    for row in data:
        try:
            if row[0][5:7] =='08':
                month_8.append(float(row[-1]))
        except:
            pass
print(month_8)
```
