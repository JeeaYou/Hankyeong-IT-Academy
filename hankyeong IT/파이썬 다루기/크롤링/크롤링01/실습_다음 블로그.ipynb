```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
검색어=input('다음 블로그 검색어 입력: ')
검색어=검색어.replace(' ','+')
url='https://search.daum.net/search?w=blog&nil_search=btn&DA=NTB&enc=utf8&q='+검색어
html=requests.get(url)
soup=bs(html.text,'html.parser')

res=[]
div_tags=soup.find_all('div','cont_inner')
for div in div_tags:
    title=div.find('a','f_link_b').text
    url=div.find('a','f_url')['href']
    date=div.find('span','f_nb date').text
    contents=div.find('p','f_eb desc').text
    writer=div.find('a','f_nb').text
#     print(div)
    
    tmp={
        '제목':title,
        '작성자':writer,
        '내용':contents,
        '작성일':date,
        'url':url
    }
    res.append(tmp)
df=pd.DataFrame(res)
display(df)
```

```python
#1~10 페이지까비 다음 블로그 크롤링하기
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

검색어=input('다음 블로그 검색어 입력: ')
검색어= 검색어.replace(' ','+')
pages=int(input('몇 페이지까지 크롤링하실래요?\n 페이지당 10개임: '))

res=[]
for page in range(1,pages+1):
    url='https://search.daum.net/search?w=blog&DA=PGD&enc=utf8&q='+검색어+'&p='+str(page)
    html=requests.get(url)
    soup=bs(html.text,'html.parser')

    div_tags=soup.find_all('div','cont_inner')
    for div in div_tags:
        try:
            title=div.find('a','f_link_b').text
            url=div.find('a','f_url')['href']
            date=div.find('span','f_nb date').text
            contents=div.find('p','f_eb desc').text
            writer=div.find('a','f_nb').text
        #     print(div)
            tmp={
                '제목':title,
                '작성자':writer,
                '내용':contents,
                '작성일':date,
                'url':url
            }
            res.append(tmp)
        except:
            pass
df=pd.DataFrame(res)
display(df)
```
