```python
from bs4 import BeautifulSoup as bs
```

```python
with open('./bs연습문제.html','r', encoding='utf-8') as f:
    soup=bs(f,'html.parser')
soup
```

### 1. img 태그만 모두 추출하시오 
-  div 태그의  클래스명이 class="flex_grid credits search_results" 아래 img태그를 모두 찾으시오(총 100개임)
- find와 find_all 활용

```python
div_tag=soup.find('div','flex_grid credits search_results')
img_tags=div_tag.find_all('img')
```

### 2. 추출한 img 태그에서 'src'속성만 모두 print하세요.

```python
for img in img_tags:
    print(img['src'])
```

### 3. 2번 결과에서 2017년도는 img_src_2017리스트에 저장하고 2018년은 img_src_2018 리스트에 저장하세요



```python
img_scr_2017=[]
img_scr_2018=[]
for img in img_tags:
    if '2017' in img['src']:
        img_scr_2017.append(img['src'])
    if '2018' in img['src']:
        img_scr_2018.append(img['src'])
display(img_scr_2017)       
display(img_scr_2018) 
```
