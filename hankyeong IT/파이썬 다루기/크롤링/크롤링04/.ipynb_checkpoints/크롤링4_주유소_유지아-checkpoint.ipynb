```python
df.columns=['상호', '가격', '상표', '셀프여부', '주소', '시']
```

```python
df_top10=df.sort_values('가격',ascending=False)[:10] #상위 10의 값 뽑기
df_top10=df_top10.reset_index(drop=True)
```

```python
df_top10['위도']=[ggcode.get_lat(i) for i in df_top10['주소']]
df_top10['경도']=[ggcode.get_lng(i) for i in df_top10['주소']]
df_top10
```

```python
df_last10=df.sort_values(by='가격',ascending=True)
df_last10=df_last10.reset_index(drop=True)
```

```python
df_last10['위도']=[ggcode.get_lat(i) for i in df_last10['주소']]
df_last10['경도']=[ggcode.get_lng(i) for i in df_last10['주소']]
df_last10
```

```python
map=folium.Map(location=[37.5610336,126.9795475],zoom_start=11) # 서울시청을 중심으로 기본차트 작성

for i in range(len(df_top10)):
    top10 = df_top10['상표'][i]
    top10 = df_top10['가격'][i]
    
    folium.Marker(
        location=[df_top10['위도'][i], df_top10['경도'][i]],
        popup=f'{상표} {가격}',   # 팝업 이쁘게 하는법
        icon=folium.Icon(color='blue',icon='cloud')
    ).add_to(map)
for i in range(len(df_top10)):
    last10 = df_last10['상표'][i]
    last10 = df_last10['가격'][i]
    
    folium.Marker(
        location=[df_last10['위도'][i], df_last10['경도'][i]],
        popup=f'{상표} {가격}',   # 팝업 이쁘게 하는법
        icon=folium.Icon(color='red',icon='cloud')
    ).add_to(map)

map
```
