```python
res=[]
for i in df1['살인']:
    min_=df1['살인'].min()
    max_=df1['살인'].max()
    result=(i-min_)/(max_-min_)
    res.append(result)
```

```python
def test(s):
    res=[]
    for i in s:
        min_=s.min()
        max_=s.max()
        result=(i-min_)/(max_-min_)
        res.append(result)
    return res
```

```python
cols=['살인','강도','강간','절도','폭력']
df1[cols]=df1[cols].apply(test)
```
