```python
plt.figure(figsize=(15,5))
df.groupby('nat')['period'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('국가별 체류기간')
```

```python
ls=['period','expense','overal']
for i in ls:
    df.groupby('nat')[i].mean().sort_values(ascending=False).plot(kind='barh')
    plt.title(f'국가별평균: {i}')
    plt.savefig(f'{i}.jpg')
    plt.show()
```

```python
names=df['nat'].unique()
ls=['period','expense','overal','cash']
for name in names:
    temp=df[df['nat']==name]
    plt.figure(figsize=(15,3))
    for idx,i in enumerate(ls):
        plt.subplot(1,len(ls),idx+1)
        plt.scatter(temp.index,temp[i])
        plt.title(f'{name}의 {i}차트')
```
