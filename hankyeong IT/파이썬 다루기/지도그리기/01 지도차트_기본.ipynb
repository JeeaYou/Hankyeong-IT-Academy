## MarkerCluster 활용 (1) -서울시 경찰서 위치

```python
df=pd.read_excel('./crim_1.xlsx')
df.shape
```

```python
#서울시청 중심 기본차트 작성
m = folium.Map(
  location=[36.5053542,127.7043419],
  zoom_start=7
)

 
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(m)
 
for i in range(len(df)):
  folium.Marker(
    location=[df.위도[i],df.경도[i]],#location=[위도,경도]
    popup=df.자치구[i],#팝업 메시지 내용은 자치구명으로 함
    icon=folium.Icon(color='red',icon='ok'),
  ).add_to(marker_cluster)
 
m.save('서울시경찰서.html')
```

## MarkerCluster 활용 (2) -전국 사망 교통사고 자도

```python
police=pd.read_csv('2016년_사망교통사고.csv', encoding='cp949')
```

```python
police.shape
```

```python
police.head(3)
```

```python
police.columns
```

```python
#발생지시도, 발생지시군구, 위도, 경도 만 필터링
police=police[['발생지시도','발생지시군구', '위도','경도']]
```

```python
police['지역']=police.발생지시도+" "+police.발생지시군구
```

```python
police.head()
```

```python
m = folium.Map(
  location=[36.5053542,127.7043419],
  zoom_start=7
)

 
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(m)
 
for i in range(len(police)):
  folium.Marker(
    location=[police.위도[i],police.경도[i]],
    popup=police.지역[i],
    icon=folium.Icon(color='red',icon='ok'),
  ).add_to(marker_cluster)
 
m.save('./map_사고.html')
```
