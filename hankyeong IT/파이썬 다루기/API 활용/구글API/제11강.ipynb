```python
ls = ['서울시청', '대구시청', '대전시청']
res = []

for v in ls:
    tmp = gmaps.geocode(v, language='ko')
    dic = {
        v: {
            'addr': tmp[0]['formatted_address'],
            'lat': tmp[0]['geometry']['location']['lat'],
            'lng': tmp[0]['geometry']['location']['lng']
        }
    }
    res.append(dic)
res
```
