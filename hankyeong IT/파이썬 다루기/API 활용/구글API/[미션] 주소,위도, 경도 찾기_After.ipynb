```python
s_addr=[]
s_lat=[]
s_lng=[]
```

```python
for i in df['자치구']:
   tmp=gmaps.geocode(i, language='ko')
   s_addr.append(tmp[0]['formatted_address'])
   s_lat.append(tmp[0]['geometry']['location']['lat'])
   s_lng.append(tmp[0]['geometry']['location']['lng'])
```
