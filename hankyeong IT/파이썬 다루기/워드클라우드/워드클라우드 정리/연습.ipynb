```python
wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
                width=300,
               height=300,
               max_words = 50,
               stopwords = stopwords)
plt.figure(figsize=(20,20))
for idx,i in enumerate(sample['text']):
    okt_nouns=okt.nouns(i)
    counter=Counter(okt_nouns)
    #워드 클라우드 완성
    wc = wc.generate_from_frequencies(counter)
    plt.subplot(1,5,idx+1)
    plt.imshow(wc)
    plt.axis('off')
```
