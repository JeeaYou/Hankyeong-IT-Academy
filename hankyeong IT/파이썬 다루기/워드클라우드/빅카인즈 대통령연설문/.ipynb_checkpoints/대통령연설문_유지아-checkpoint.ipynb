```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs 
```

```python
url='https://www.korea.kr/archive/speechView.do?newsId=132033493'
html=requests.get(url)
soup=bs(html.text,'html.parser')
```

```python
 txt=soup.find('div','article-content').text
```

```python
from konlpy.tag import Okt
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rc('axes', unicode_minus=False)
```

# 1. 워드 클라우드 

```python
okt=Okt()
re=okt.nouns(txt)
re1=' '.join(re)
re1
wc = WordCloud('./NanumGothic.ttf',
               background_color = "white",
               max_words = 2000)
wc = wc.generate(re1)
plt.figure(figsize = (12, 12))
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
from collections import Counter
```

```python
re2=re1.split(' ') 
cnt = Counter(re2)
word=cnt.most_common(50)
df=pd.DataFrame(word, columns=['단어','횟수'])
display(df)
```

#  2. 단어 차트화

```python
df.set_index('단어').plot(kind='bar', figsize=(15,5))
plt.grid()
```
