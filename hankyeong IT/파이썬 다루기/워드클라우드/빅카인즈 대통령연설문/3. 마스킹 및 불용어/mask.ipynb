### 워드클라우드에 마스크 적용
#### png로 저장된 이미지자료와
#### 텍스트 글자를 합쳐서 표현

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# %matplotlib inline
#plt.rcParams['font.family'] = 'NanumGothic'
```

```python
f=open('./data/txt_sample1.txt','r')
lines=f.read()
f.close
```

```python
from PIL import Image
import numpy as np

#mask_[4]
#print(mask_[1])
```

```python
mask_=np.array(Image.open('aaa.png'))
```

```python
def myWord(color,pic):
    mask_=np.array(Image.open(pic))
    wc=WordCloud(font_path='./NanumGothic.ttf',
                 background_color=color,
                 width=1000,height=1000,
                 max_words=100,
                 max_font_size=300,
                mask=mask_)
    wordFre=wc.generate(lines)

    plt.figure(figsize=(12,12))
    plt.imshow(wordFre, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
# 만약에 사진리스트가 있는 있는 폴더가 있다면
# 폴더내의 자료를 모두 읽어서
# 그 자료를 for로 돌리게 할수 있음.
# 그러면 아래의 li= 직접 파일명을 입력하지 않아도 됨

li=['star.png','dog.png','dog.jpg']  #stat.png / dog.png / dog.jpg는 작업폴더에 있어야함.

for i in li:
    myWord('white',i)
```
