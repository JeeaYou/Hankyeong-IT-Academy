```python
txtList=['강아지 산책 강아지 목욕 강아지 미용 강아지 쇼핑',
 '친구와 저녁 먹음 가족과 점심 먹음 혼자 저녁 먹음 친구와 쇼핑']
txt=' '.join(txtList)
find_word=['강아지','미용','친구']
cnt=[txt.count(i) for i in find_word]
plt.bar(find_word,cnt)
```

```python
txt_sum=[j/sum(cnt) for j in cnt]
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.bar(find_word,cnt)
plt.title('워드 빈도수 막대그래프')
plt.ylim(0,10)

plt.subplot(132)
plt.bar(find_word,txt_sum)
plt.title('워드 빈도수 막대그래프%')

plt.subplot(133)
plt.pie(cnt,labels=find_word,autopct='%1.1f%%')   # %나오게 만들기
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import matplotlib.pyplot as plt
```

```python
url='https://movie.naver.com/movie/bi/mi/basic.naver?code=190991'
response=requests.get(url)

if response.status_code==200:
    html=response.text
    soup=bs(html,'html.parser')
else:
    print(response,status_code)
    
txt=soup.find_all('p','con_tx')
txtl=[i.text for i in txt]
text_list=''.join(txtl)

word=['수학','학문','과학']
cnt=[text_list.count(j) for j in word]

plt.figure(figsize=(15,6))
plt.subplot(131)
plt.title('막대그래프')
plt.bar(word,cnt)

percent=[a/sum(cnt) for a in cnt]
plt.subplot(132)
plt.title('퍼센트 그래프%')
plt.bar(word,percent)

plt.subplot(133)
plt.title('원형차트')
plt.pie(cnt,labels=word,autopct='%.1f%%',counterclock=False)
plt.show()
```
