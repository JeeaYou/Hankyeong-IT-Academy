## 백내장 데이터 증강 코드


```python
# 백내장 데이터 증강 코드
import time
import os
from PIL import Image
import sys
from glob import glob
from matplotlib.pyplot import imshow, subplots, title
from torchvision import transforms
import albumentations
import random
 
out_dir ="cataract_eyes"
if out_dir not in os.listdir():
    os.mkdir(out_dir)

def image(path):
    print("Process Start.")
    start_time = time.time()

    # 결과 저장 폴더 생성
    image = Image.open(path)
    Xdim, Ydim = image.size
    COUNT = 1

    # 원본파일
    temp_new_file_name = path.split('/')[-1].split('.')[0].split('\\')[-1]+'_'+"%05d.jpg" %COUNT
    print(temp_new_file_name)

    COUNT += 1

    image.save(out_dir + "/" + temp_new_file_name)
    image.close()

    FILELIST = [temp_new_file_name]

    # 폴더 내의 이미지를 모두 읽어와 좌우대칭을 저장합니다. 2의 1승
    for i in range(len(FILELIST)):
        image = Image.open(path)
        new_temp_name = path.split('/')[-1].split('.')[0].split('\\')[-1]+'_좌우_'+"%05d.jpg" %COUNT
        COUNT += 1

        image = image.transpose(Image.FLIP_LEFT_RIGHT)
        image.save(out_dir + "/" + new_temp_name)
        image.close()

        FILELIST.append(new_temp_name)
    
    # 명암조정 데이터
    for i in range(len(FILELIST)):
        image = Image.open(out_dir + "/" + FILELIST[i])

        COUNT += 1

        loader_transform1 = transforms.ColorJitter(brightness=2)
        loader_transform2 = transforms.ColorJitter(contrast=2)
        loader_transform3 = transforms.ColorJitter(saturation=2)

        img_aug1 = loader_transform1(image)
        img_aug2 = loader_transform2(image)
        img_aug3 = loader_transform3(image)

        img_aug1.save(out_dir + "/" + path.split('/')[-1].split('.')[0].split('\\')[-1]+'_명암_1_'+"%05d.jpg" %COUNT)
        img_aug2.save(out_dir + "/" + path.split('/')[-1].split('.')[0].split('\\')[-1]+'_명암_2_'+"%05d.jpg" %COUNT)
        img_aug3.save(out_dir + "/" + path.split('/')[-1].split('.')[0].split('\\')[-1]+'_명암_3_'+"%05d.jpg" %COUNT)

        image.close()

    # 리스트 안의 이미지를 모두 읽어와 1도씩 회전합니다. 2의 3승 * 180
    for el in FILELIST:
        for i in range(20):
            if COUNT > 1000:
                break
            image = Image.open(out_dir + "/" + el)
            new_temp_name = out_dir + "/" + path.split('/')[-1].split('.')[0].split('\\')[-1]+'_회전_'+"%05d.jpg" %COUNT
            
            COUNT += 1
            image = image.rotate(i+1)
            image = image.resize((Xdim, Ydim))
            image.save( new_temp_name)
            image.close()
    end_time = time.time()
    print("The Job Took " + str(end_time - start_time) + " seconds.")
```

```python
import random
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps    
from glob import glob
```

```python
#다음 변수를 수정하여 새로 만들 이미지 갯수를 정합니다.
num_augmented_images = 453

file_path = './data/cataract원본/'
file_names = os.listdir(file_path)
total_origin_image_num = len(file_names)
augment_cnt = 1

for i in range(1, num_augmented_images):
    change_picture_index = random.randrange(1, total_origin_image_num-1)
    print(change_picture_index)
    print(file_names[change_picture_index])
    file_name = file_names[change_picture_index]
    
    origin_image_path = './data/cataract원본/' + file_name
    print(origin_image_path)
    image = Image.open(origin_image_path)
    random_augment = random.randrange(1,4)
    
    if(random_augment == 1):
        #이미지 좌우 반전
        print("invert")
        inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
        inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.jpg')
        
    elif(random_augment == 2):
        #이미지 기울이기
        print("rotate")
        rotated_image = image.rotate(random.randrange(-20, 20))
        rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.jpg')
        
    elif(random_augment == 3):
        #노이즈 추가하기
        img = cv2.imread(origin_image_path)
        print("noise")
        row,col,ch= img.shape
        mean = 0
        var = 0.1
        sigma = var**0.5
        gauss = np.random.normal(mean,sigma,(row,col,ch))
        gauss = gauss.reshape(row,col,ch)
        noisy_array = img + gauss
        noisy_image = Image.fromarray(np.uint8(noisy_array)).convert('RGB')
        noisy_image.save(file_path + 'noiseAdded_' + str(augment_cnt) + '.jpg')
        
    augment_cnt += 1
```
