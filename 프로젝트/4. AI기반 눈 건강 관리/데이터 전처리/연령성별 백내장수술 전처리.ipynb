```python
import pandas as pd
import matplotlib.pyplot as plt   
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./파일데이터/연령별_성별_백내장수술_현황_계__20221202113104.csv',encoding='cp949')
df1 = df.copy()
df1.head(10)
```

```python
df1 = df1[['연령별(1)', '성별(1)', '2011', '2011.1',  '2012', '2012.1',
       '2013', '2013.1',  '2014',
       '2014.1', '2015', '2015.1',
       '2016', '2016.1', '2017', '2017.1', '2018', '2018.1', '2019', '2019.1',
       '2020', '2020.1']]
df1 = df1[1:]
df1 = df1.T.reset_index().T
df1.head(10)
```

```python
cols = df1[:2]
cols = cols.T
cols.columns = ['연도별','항목']
cols['total'] = cols['연도별'] + '/'+ cols['항목']
cols_total = cols['total'].values
cols_total
```

```python
df2 = df1[2:]
df2.columns = cols_total
df2.columns = ['연령별', '성별', '2011/수술인원 (명)', '2011.1/수술건수 (건)',
       '2012/수술인원 (명)', '2012.1/수술건수 (건)', '2013/수술인원 (명)', '2013.1/수술건수 (건)',
       '2014/수술인원 (명)', '2014.1/수술건수 (건)', '2015/수술인원 (명)', '2015.1/수술건수 (건)',
       '2016/수술인원 (명)', '2016.1/수술건수 (건)', '2017/수술인원 (명)', '2017.1/수술건수 (건)',
       '2018/수술인원 (명)', '2018.1/수술건수 (건)', '2019/수술인원 (명)', '2019.1/수술건수 (건)',
       '2020/수술인원 (명)', '2020.1/수술건수 (건)']
df2 = df2.set_index(['연령별','성별'])
df2.head(10)
```

```python
df3 = df2.stack().reset_index()
df3['년도'] = [i.split('/')[0] for i in df3['level_2']]
df3['항목'] = [i.split('/')[1] for i in df3['level_2']]
del df3['level_2']
df3.columns = ['연령별', '성별', '값', '년도', '항목']
df3 = df3.set_index(['년도', '성별' , '항목','연령별']).unstack().reset_index()
df3['년도'] = df3['년도'].astype(float).astype('int')
df3 = df3.set_index(['년도','성별','항목'])
df3 = df3.T.reset_index()
df3.head(10)
```

```python
df_finall = df3.T.rename(columns=df3['연령별'])[2:]
df_finall['19세이하'] = df_finall.iloc[:,:4].astype(int).sum(axis=1)
df_finall = df_finall[['19세이하', '20 - 24세', '25 - 29세',
       '30 - 34세', '35 - 39세', '40 - 44세', '45 - 49세', '5 - 9세', '50 - 54세',
       '55 - 59세', '60 - 64세', '65 - 69세', '70 - 74세', '75 - 79세', '80 - 84세',
       '85세 이상', '합계']]
df_finall = df_finall[['합계','19세이하', '20 - 24세', '25 - 29세', '30 - 34세', '35 - 39세', '40 - 44세',
       '45 - 49세',  '50 - 54세', '55 - 59세', '60 - 64세', '65 - 69세',
       '70 - 74세', '75 - 79세', '80 - 84세', '85세 이상']]
df_finall.columns = ['합계','19세이하', '20 - 24세', '25 - 29세', '30 - 34세', '35 - 39세', '40 - 44세',
       '45 - 49세',  '50 - 54세', '55 - 59세', '60 - 64세', '65 - 69세',
       '70 - 74세', '75 - 79세', '80 - 84세', '85세이상']
df_finall.head()
```

```python
cataract = pd.read_csv('./전처리파일/연령별 백내장수술.csv')
cataract.head(10)
```

```python
cata_total = cataract[cataract['성별']=='합계'].reset_index(drop=True)
cata_total.head()
```

```python
total_1 = cata_total[cata_total['항목']=='수술인원 (명)']
total_2 = cata_total[cata_total['항목']=='수술건수 (건)']
```

```python
# 백내장 합계의 차트 ---- 수술건수 > 수술인원수 == 백내장 재발확률이 높다.
plt.figure(figsize=(10,6))
plt.plot(total_1['년도'],total_1['합계'],'ro--',label='수술인원 (명)')
plt.plot(total_2['년도'],total_2['합계'],'bo--',label='수술건수 (건)')
plt.xlabel('년도')
plt.ylabel('건수/인원수')
plt.grid(True)
plt.legend()

plt.show()
```

```python
cata_2020 = cataract[cataract['년도']==2020]
cata_2020 = cata_2020[cata_2020['성별']!='합계']
cata_2020
```

```python
for many in cata_2020['항목'].unique():
    plt.figure(figsize=(8,6))
    tmp = cata_2020[cata_2020['항목']==many]
    tmp = tmp.T[4:].reset_index()
    tmp.columns = ['연령별','남자','여자']
    sns.barplot(x='남자', y='연령별', data=tmp, palette='Blues')
    plt.title(many+'남자')
plt.show()
```

```python
for many in cata_2020['항목'].unique():
    plt.figure(figsize=(8,6))
    tmp = cata_2020[cata_2020['항목']==many]
    tmp = tmp.T[4:].reset_index()
    tmp.columns = ['연령별','남자','여자']
    sns.barplot(x='여자', y='연령별', data=tmp, palette='Reds')
    plt.title(many+'여자')
```

```python
# pie 차트 2020년의 데이터를 보면 여자 수술인원과 재발가능성이 더 높다.
colors = ['#d395d0','#ff9999']
explode = [0.05, 0.05]
for many in cata_2020['항목'].unique():
    plt.figure(figsize=(8,6))
    tmp = cata_2020[cata_2020['항목']==many]
    # texts, autotexts 인자를 활용하여 텍스트 스타일링을 적용합니다
    patches, texts, autotexts = plt.pie(tmp['합계'], 
                                        labels=tmp['성별'],  
                                        explode=explode,
                                        autopct='%1.1f%%',
                                        shadow=True, 
                                        colors=colors,
                                        startangle=90)
    plt.title(many, fontsize=15)
    
    # label 텍스트에 대한 스타일 적용
    for t in texts:
        t.set_fontsize(12)
        t.set_color('black')
        
    # pie 위의 텍스트에 대한 스타일 적용
    for t in autotexts:
        t.set_color("black")
        t.set_fontsize(18)

plt.show()
```

```python
cata_people = cataract[cataract['항목']=='수술인원 (명)']
del cata_people['항목']
cata_people = cata_people.set_index(['년도','성별'])
cata_people.iloc[:,:1].head(10)
```

```python
total = pd.read_csv('./전처리파일/총인구수.csv')
total = total.set_index(['년도','성별'])
total.iloc[:,:1].head(10)
```

```python
merge = pd.merge(cata_people.iloc[:,:1],total.iloc[:,:1],
         left_index=True, right_index=True,
         how='left')
merge.columns = ['백내장 환자수', '총인구수']
merge['비율 %'] = (merge['백내장 환자수'] / merge['총인구수'] * 100).round(2)
merge.head(10)
```

```python
merge_1 = merge.reset_index().set_index('년도')
merge_1 = merge_1[merge_1['성별']!='합계'].reset_index()
merge_1
```

```python
# 백내장 합계의 차트 ---- 여자의 발병건수는 남자보다 많다
merge_female = merge_1[merge_1['성별']=='여자']
merge_male = merge_1[merge_1['성별']=='남자']
plt.plot(merge_female['년도'],merge_female['비율 %'],'ro--',label='여자')
plt.plot(merge_male['년도'],merge_male['비율 %'],'bo--',label='남자')
plt.xlabel('년도')
plt.ylabel('백내장환자의 비율 %')
plt.grid(True)
plt.legend()
plt.show()
```

```python
merge_2020 = pd.merge(cata_people[-3:],total[-6:-3],
         left_index=True, right_index=True,
         how='left')
merge_2020.iloc[:,0]
```

```python
merge_2020.iloc[:,16]
```

```python
col_2020 = ['합계', '19세이하', '20 - 24세', '25 - 29세', '30 - 34세', '35 - 39세',
       '40 - 44세', '45 - 49세', '50 - 54세', '55 - 59세', '60 - 64세', '65 - 69세',
       '70 - 74세', '75 - 79세', '80 - 84세', '85세이상']
cnt = 0
merge_2020_copy = merge_2020.copy()
for col in col_2020:
    merge_2020_copy[col] = (merge_2020_copy.iloc[:,cnt] / merge_2020_copy.iloc[:,cnt+16] * 100).round(2)
    cnt+=1
merge_2020_per = merge_2020_copy.iloc[:,-15:]
merge_2020_per
```

```python
merge_2020_per = merge_2020_per.reset_index().iloc[:,1:].set_index('성별').T
merge_2020_per = merge_2020_per.reset_index()
```

```python
merge_2020_per.columns = ['연령별 비율%', '남자', '여자', '합계']
merge_2020_per
```

```python
# 남자 -- 백내장 발병나이는 60세이상, 다발나이는 70세-84세
x = merge_2020_per['연령별 비율%']
y = merge_2020_per['남자']
plt.figure(figsize=(14,6))
plt.plot(x,y )
plt.xlabel('연령별')
plt.ylabel('백내장환자 비율% (남자)')
plt.fill_between(x[9:], y[9:], alpha=0.5)
plt.grid()
plt.show()
```

```python
# 여자 -- 백내장 발병나이는 55세이상, 다발나이는 70세-79세
x = merge_2020_per['연령별 비율%']
y = merge_2020_per['여자']
plt.figure(figsize=(14,6))
plt.plot(x,y )
plt.xlabel('연령별')
plt.ylabel('백내장환자 비율% (여자)')
plt.fill_between(x[8:], y[8:], alpha=0.5)
plt.grid()
plt.show()
```

```python
# 합계 -- 평균 백내장 발병나이는 60세이상, 여자의 백내장 발병나이가 남자보다 빠르다
x = merge_2020_per['연령별 비율%']
y = merge_2020_per['합계']
plt.figure(figsize=(14,6))
plt.plot(x,y )
plt.xlabel('연령별')
plt.ylabel('백내장환자 비율% (합계)')
plt.fill_between(x[9:], y[9:], alpha=0.5)
plt.grid()
plt.show()
```
