```python
import pandas as pd
import matplotlib.pyplot as plt   
import numpy as np
```

```python
df = pd.read_csv('./파일데이터/주민등록총인구수_20221202173252.csv',encoding='cp949')
df
```

```python
df_total = df.transpose()
df_total = df_total.drop(['행정구역(시군구)별'])
df_total = df_total.rename(columns=df_total.iloc[0]).drop(['연령별'])
df_total['19세이하'] = df_total.iloc[:,22:27].astype(int).sum(axis=1)
df_total.head(10)
```

```python
df_total.iloc[:,22:27].head()
# df_total.iloc[:,27:32]
# df_total.iloc[:,32:37].head()
```

```python
cnt = 22
for x in df_total:
    df_total[f'{cnt-2} - {cnt+2}세'] = df_total.iloc[:,cnt:cnt+5].astype(int).sum(axis=1)
    cnt+=5
    if cnt >=84:
        break
df_total.head(10)
```

```python
df_total['85세이상'] = df_total.iloc[:,87:103].astype(int).sum(axis=1)
df_total.head(10)
```

```python
df_total1 = df_total.copy()
df_total1 = df_total1[['연령별', '계', '19세이하', '20 - 24세', '25 - 29세', '30 - 34세',
       '35 - 39세', '40 - 44세', '45 - 49세', '50 - 54세', '55 - 59세', '60 - 64세',
       '65 - 69세', '70 - 74세', '75 - 79세', '80 - 84세', '85세이상']]
df_total1 = df_total1.reset_index()
df_total1['index'] = df_total1['index'].astype(float).astype(int)
df_total1.columns = ['년도', '성별', '합계',  '19세이하', '20 - 24세', '25 - 29세', '30 - 34세',
       '35 - 39세', '40 - 44세', '45 - 49세', '50 - 54세', '55 - 59세', '60 - 64세',
       '65 - 69세', '70 - 74세', '75 - 79세', '80 - 84세', '85세이상']
df_total1 = df_total1.set_index(['년도'])
df_total1['성별'] = df_total1['성별'].replace('총인구수 (명)','합계').replace('남자인구수 (명)','남자').replace('여자인구수 (명)','여자')
df_total1[df_total1['성별']=='합계']
df_total1.head(10)
```

```python
df_total1.to_csv('./전처리파일/총인구수.csv')
```
