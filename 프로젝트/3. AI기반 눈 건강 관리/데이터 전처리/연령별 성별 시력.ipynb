```python
import pandas as pd
import matplotlib.pyplot as plt   
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./파일데이터/연령별_시력_분포_현황_20221202113339.csv',encoding='cp949')
df.head(10)
```

```python
df['연령별 성별'] = df['연령별(1)'] + '/' + df['성별(1)']
col = df['연령별 성별']
col.head(10)
```

```python
df1 = df.T
df1.columns = col
df1 = df1.reset_index()
df1.columns = ['년도', '눈', '시력', '합계/합계', '남자/남자', '여자/여자',
       '19세 이하/합계', '19세 이하/남자', '19세 이하/여자', '20 ~ 24세/합계', '20 ~ 24세/남자',
       '20 ~ 24세/여자', '25 ~ 29세/합계', '25 ~ 29세/남자', '25 ~ 29세/여자',
       '30 ~ 34세/합계', '30 ~ 34세/남자', '30 ~ 34세/여자', '35 ~ 39세/합계',
       '35 ~ 39세/남자', '35 ~ 39세/여자', '40 ~ 44세/합계', '40 ~ 44세/남자',
       '40 ~ 44세/여자', '45 ~ 49세/합계', '45 ~ 49세/남자', '45 ~ 49세/여자',
       '50 ~ 54세/합계', '50 ~ 54세/남자', '50 ~ 54세/여자', '55 ~ 59세/합계',
       '55 ~ 59세/남자', '55 ~ 59세/여자', '60 ~ 64세/합계', '60 ~ 64세/남자',
       '60 ~ 64세/여자', '65 ~ 69세/합계', '65 ~ 69세/남자', '65 ~ 69세/여자',
       '70 ~ 74세/합계', '70 ~ 74세/남자', '70 ~ 74세/여자', '75 ~ 79세/합계',
       '75 ~ 79세/남자', '75 ~ 79세/여자', '80 ~ 84세/합계', '80 ~ 84세/남자',
       '80 ~ 84세/여자', '85세 이상/합계', '85세 이상/남자', '85세 이상/여자']
df1.head(10)
```

```python
df2 = df1[2:].set_index(['년도', '눈', '시력'])
df2 = df2.stack().reset_index()
df2['연령별'] = [i.split('/')[0] for i in df2['level_3']]
df2['성별'] = [i.split('/')[1] for i in df2['level_3']]
del df2['level_3']
df2.columns = ['년도', '눈', '시력', '값', '연령별', '성별']
df2.head(10)
```

```python
df3 = df2.set_index(['년도','시력', '눈','성별', '연령별' ]).unstack().reset_index()
df3 = df3.iloc[:-3,:-3]
df3['년도'] = df3['년도'].astype(float).astype('int')
df3['눈'] = [x.replace('안 시력','') for x in df3['눈']]
df3.head(10)
```

```python
df4 = df3.T.reset_index()
df4 = df4.set_index('연령별').T[1:]
df4.columns = ['년도', '시력', '눈', '성별', '19세이하', '20 - 24세', '25 - 29세', '30 - 34세',
       '35 - 39세', '40 - 44세', '45 - 49세', '50 - 54세', '55 - 59세', '60 - 64세',
       '65 - 69세', '70 - 74세', '75 - 79세', '80 - 84세', '85세이상']
df4 = df4.set_index(['년도','시력','눈','성별'])
df4.head(10)
```

```python
df4['합계'] = df4[:].replace('-','0').astype(int).sum(axis=1)
df4 = df4[['합계','19세이하', '20 - 24세', '25 - 29세', '30 - 34세', '35 - 39세', '40 - 44세',
       '45 - 49세', '50 - 54세', '55 - 59세', '60 - 64세', '65 - 69세', '70 - 74세',
       '75 - 79세', '80 - 84세', '85세이상']]
df4
```

```python
df4.to_csv('./전처리파일/연령별시력.csv')
```
